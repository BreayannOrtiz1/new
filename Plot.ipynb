```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,11)
y1 = x**2.0
y2 = x**1.9

plt.plot(x,y2,'gs-',y1,'bo-',markersize=7);
plt.grid()
```
